```python
#class,Constructor

class Person:
    #constructor
    def __init__(self,_name,_age):
        self.name = _name
        self.age = _age
        self.planet = "Earth"
    def __str__(self):
        return f"name: {self.name}, age: {self.age}, planet: {self.planet}"  

    # annotation
    @staticmethod
    def print():
        print("Hello")  
```

```python
person = Person("name", 20)
person2 = Person("name", 20)
```

```python
print(person)
print(person2)
```

```python
#inheritance
class Employee(Person):
  def __init__(self, name, age, id):
    super().__init__(name, age)
    self.id = id
```

```python
e = Employee("A", 20, "John Jr")
print(e)
print(e.name, e.age, e.id)
```

```python
isinstance(e,Person)
```

```python
issubclass(Employee,Person)
```
